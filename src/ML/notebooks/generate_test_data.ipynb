```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Установка зерна для воспроизводимости
np.random.seed(123)

# Генерация дат
dates = pd.date_range(start='2023-06-23', periods=365, freq='D')

# Стартовая цена и список
prices = []
current_price = 1599

for _ in range(len(dates)):
    if np.random.rand() < 0.7:
        delta = np.random.choice([0, 10, -10, 20, -20, 0, 0])
    else:
        delta = np.random.choice([-50, 50])
    
    current_price = max(990, current_price + delta)
    prices.append(current_price)

# Создание и сохранение DataFrame
alt_df = pd.DataFrame({'date': dates, 'price': prices})
alt_df.to_csv('final_test_price_data.csv', index=False)
print("✅ Тестовый файл сохранён как final_test_price_data.csv")

# Визуализация
plt.figure(figsize=(14, 6))
plt.plot(alt_df['date'], alt_df['price'], label='Цена', color='blue')
plt.title('🟦 Альтернативная история цен (365 дней)')
plt.xlabel('Дата')
plt.ylabel('Цена (₽)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
