```python
import pandas as pd
import numpy as np
import random

# Установка сидов для воспроизводимости
random.seed(42)
np.random.seed(42)

# Кол-во дней
num_days = 365
base_price = 1500

dates = pd.date_range(end=pd.Timestamp.today(), periods=num_days)
prices = []

current_price = base_price
for i in range(num_days):
    # С вероятностью 70% цена остается прежней
    if random.random() < 0.7:
        prices.append(current_price)
    else:
        # Иначе случайное изменение в пределах ±2.5%
        change = random.uniform(-0.025, 0.025)
        current_price = round(current_price * (1 + change), 2)
        prices.append(current_price)

# Создаем DataFrame
df = pd.DataFrame({
    'date': dates,
    'price': prices
})

# Сохраняем в CSV
df.to_csv('../mock_price_data.csv', index=False)

print("✅ Данные на 365 дней сохранены в mock_price_data.csv")
df.tail()
```
